```python
from nltk.corpus import wordnet

def get_antonyms(word):
    antonyms = set()
    for syn in wordnet.synsets(word):
        for lemma in syn.lemmas():
            for antonym in lemma.antonyms():
                antonyms.add(antonym.name())
    return antonyms
```

```python
words = ["good", "happy", "fast","sad","slow"]
for word in words:
  print(f"Antonyms of '{word}': {get_antonyms(word)}")
```

```python
from nltk.stem import SnowballStemmer

language = "spanish"
words = ["corriendo", "comer", "gatos", "bibliotecas", "universidades"]
stemmer = SnowballStemmer(language)
stemmed_words = [stemmer.stem(word) for word in words]

for original, stemmed in zip(words, stemmed_words):
    print(f"Original: {original}, Stemmed: {stemmed}")
```

```python
words2 = ["running", "happiness", "faster","crying","slower"]
from nltk.stem import  PorterStemmer, LancasterStemmer, SnowballStemmer

nltk.download('wordnet')

def stem(word):
    pStemmer = PorterStemmer()
    lStemmer = LancasterStemmer()
    sStemmer = SnowballStemmer(language='english')
    print("Wors : ",word)
    print("Porter Stemmet : ",pStemmer.stem(word))
    print("Lancaster Stemmet : ",lStemmer.stem(word))
    print("Snowball Stemmet : ",sStemmer.stem(word))

for word in words2:
    stem(word)
```

```python
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer

nltk.download('wordnet')

def stem_word(word):
    porter_stemmer = PorterStemmer()
    stemmed_word = porter_stemmer.stem(word)
    return stemmed_word

def lemmatize_word(word):
    lemmatizer = WordNetLemmatizer()
    lemmatized_word = lemmatizer.lemmatize(word)
    return lemmatized_word

words = ["running", "flies", "running", "happiness", "dogs"]


for word in words:
    stemmed_word = stem_word(word)
    lemmatized_word = lemmatize_word(word)
    print(f"{word} {stemmed_word} {lemmatized_word}")
```

```python
text = "Hi, my name is Sanin Zulphi. I am 22 years old and I am from Kerala, Malappuram."
```

```python
from nltk import word_tokenize
from nltk.corpus import stopwords
import string
import nltk

nltk.download('punkt')
nltk.download('stopwords')

def preprocess_text(text):
    wt = word_tokenize(text)
    wt = [word for word in wt if word not in string.punctuation]
    stop_words = set(stopwords.words('english'))
    wt = [word for word in wt if word.lower() not in stop_words]
    return wt

cleanText = preprocess_text(text)
print(cleanText)
```

```python
from nltk import pos_tag
nltk.download('averaged_perceptron_tagger')
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag

pos_tags = pos_tag(cleanText)

print("\nPoS Tagging Result:")
for word, pos_tag in pos_tags:
	print(f"{word}: {pos_tag}")
```

```python
import matplotlib.pyplot as plt
from collections import Counter
pos_tag_counts = Counter(tag[1] for tag in pos_tags)

# Extract POS tag labels and counts for plotting
labels, counts = zip(*pos_tag_counts.items())

# Plotting the distribution
plt.figure(figsize=(13,10))
plt.bar(labels, counts, color='blue')
plt.title('POS Tag Distribution')
plt.xlabel('POS Tags')
plt.ylabel('Frequency')
plt.show()
```
